# Modelling (Random Forest)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
import pandas as pd


columns_to_impute = [
    'COMM_MEMBER', 'COMM_ORG', 'PATIENTS', 'CLINICIANS',
    'PROF_ASSOC', 'POLICY_MAKERS', 'GOVT_REPS'
]


X = df.drop(columns=['FINAL_PAR', 'EMPLOYMENT'], errors='ignore')
y = df['EMPLOYMENT']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

for col in columns_to_impute:
    mode_value = X_train[col].mode()[0]
    X_train[col].fillna(mode_value, inplace=True)
    X_test[col].fillna(mode_value, inplace=True)


pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('random_forest', RandomForestClassifier(
        criterion='entropy',
        class_weight='balanced',
        oob_score=True,
        random_state=42
    ))
])


param_dist = {
    "random_forest__n_estimators": [200],
    "random_forest__max_depth": [5, 10, 15],
    "random_forest__min_samples_leaf": [15, 20, 25],
    "random_forest__max_features": ['sqrt']
}

skf = StratifiedKFold(n_splits=5)

random_search = GridSearchCV(pipeline, param_grid=param_dist, cv=skf)
random_search.fit(X_train, y_train)


results = pd.DataFrame(random_search.cv_results_)
best_results = results[results['rank_test_score'] == 1]


print("Best Hyperparameters:", random_search.best_params_)
print("Best Model Accuracy:", random_search.best_score_)
```

```python
from sklearn.metrics import accuracy_score


best_params = random_search.best_params_


best_model = Pipeline([
    ('scaler', StandardScaler()),
    ('random_forest', RandomForestClassifier(
        criterion='entropy',
        class_weight='balanced',
        oob_score=True,
        random_state=42,
        n_estimators=best_params['random_forest__n_estimators'],
        max_depth=best_params['random_forest__max_depth'],
        min_samples_leaf=best_params['random_forest__min_samples_leaf'],
        max_features=best_params['random_forest__max_features']
    ))
])


best_model.fit(X_train, y_train)


y_train_pred = best_model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)


y_test_pred = best_model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)


print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report


y_test_pred = best_model.predict(X_test)


conf_matrix = confusion_matrix(y_test, y_test_pred)


print("Confusion Matrix:")
print(conf_matrix)


print("\nClassification Report:")
print(classification_report(y_test, y_test_pred))
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score


columns_to_impute = [
    'COMM_MEMBER', 'COMM_ORG', 'PATIENTS', 'CLINICIANS',
    'PROF_ASSOC', 'POLICY_MAKERS', 'GOVT_REPS'
]


X = df.drop(columns=['FINAL_PAR', 'EMPLOYMENT'], errors='ignore')
y = df['EMPLOYMENT']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

for col in columns_to_impute:
    mode_value = X_train[col].mode()[0]
    X_train[col] = X_train[col].fillna(mode_value)
    X_test[col] = X_test[col].fillna(mode_value)


print("Missing values before SMOTE:", X_train.isna().sum().sum())


smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)


scale_pos_weight = y_train.value_counts()[0] / y_train.value_counts()[1]


xgb_model = XGBClassifier(
    objective='binary:logistic',
    scale_pos_weight=scale_pos_weight,
    eval_metric='auc',
    use_label_encoder=False,
    random_state=42
)

xgb_model.fit(X_train_resampled, y_train_resampled)

y_probs = xgb_model.predict_proba(X_test)[:, 1]
y_pred = (y_probs >= 0.5).astype(int)

print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_probs))
```

# KMEANS

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import silhouette_score
import pandas as pd


columns_to_impute = [
    'COMM_MEMBER', 'COMM_ORG', 'PATIENTS', 'CLINICIANS',
    'PROF_ASSOC', 'POLICY_MAKERS', 'GOVT_REPS'
]

X = df.drop(columns=['FINAL_PAR', 'EMPLOYMENT'], errors='ignore')

for col in columns_to_impute:
    mode_value = X[col].mode()[0]
    X[col].fillna(mode_value, inplace=True)

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('kmeans', KMeans(random_state=42, n_init=10))
])


param_grid = {
    "kmeans__n_clusters": [2, 3, 4, 5, 6, 7, 8, 9, 10]  # Testing different K values
}


best_score = -1
best_k = None
best_model = None

for k in param_grid["kmeans__n_clusters"]:
    model = Pipeline([
        ('scaler', StandardScaler()),
        ('kmeans', KMeans(n_clusters=k, random_state=42, n_init=10))
    ])

    model.fit(X)
    labels = model.named_steps['kmeans'].labels_
    score = silhouette_score(X, labels)

    print(f'K={k}, Silhouette Score={score:.4f}')

    if score > best_score:
        best_score = score
        best_k = k
        best_model = model


print("Best number of clusters (K):", best_k)
print("Best Silhouette Score:", best_score)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

kmeans = KMeans(n_clusters=best_k, random_state=42)
clusters = kmeans.fit_predict(X_pca)


pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Cluster'] = clusters
pca_df['EMPLOYMENT'] = y.values


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', palette='viridis', data=pca_df, alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering (PCA Reduced)')
plt.legend()


plt.subplot(1, 2, 2)
sns.scatterplot(x='PC1', y='PC2', hue='EMPLOYMENT', palette='tab10', data=pca_df, alpha=0.7)
plt.title('Clusters Based on EMPLOYMENT')
plt.legend()

plt.tight_layout()
plt.show()
```
